```python
# similarity_retrieval/vector_search.py

from langchain_chroma import Chroma
from models.embedding_models import embed_model


def vector_search(user_query: str, top_k: int = 5):
    """Searches the persistent Chroma vector database using similarity retrieval."""
    db = Chroma(
        persist_directory="./data_warehouse/rag_vector_db",
        collection_name="rag_collection",
        embedding_function=embed_model,
    )

    retriever = db.as_retriever(
        search_type="similarity", search_kwargs={"k": top_k}
    )

    return retriever.invoke(user_query)
```

```python

from langchain_chroma import Chroma
from models.embedding_models import embed_model
def load_vector_db():
    """Loads the persistent Chroma vector database."""
    return Chroma(
        persist_directory="data_warehouse/rag_vector_db",
        collection_name="rag_collection",
        embedding_function=embed_model,
    )


def vector_search(
        user_query: str,
        top_k: int = 5
):

    print("Loading DB...")

    db = load_vector_db()

    print("Creating retriever...")

    retriever = db.as_retriever(
        search_type="similarity",
        search_kwargs={"k": top_k}
    )

    print("Retriever created")

    print("Invoking retriever...")

    docs = retriever.invoke(user_query)

    print("Retrieved docs")

    return docs


# retrieved_chunks = vector_search(
#     user_query="what is python",
#     top_k=5
# )
```
